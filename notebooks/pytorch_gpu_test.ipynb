```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a simple dataset
data = {'x': [1, 2, 3, 4, 5], 'y': [10, 15, 7, 10, 25]}
df = pd.DataFrame(data)

# Plot the data
plt.plot(df['x'], df['y'], marker='o')
plt.title('Simple Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```
